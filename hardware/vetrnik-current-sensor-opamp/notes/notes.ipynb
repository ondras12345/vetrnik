## Calibration
- ACS770 connected to output of KORAD KA3005P
- opamp output cannot go below ~50mV (measured), so Vref shouldn't be tweaked to 500mV
- the output showed parasitic oscillation, so a 22pF capacitor was added
- 440mV is configured as `$3=21` `$4=0` `$2=41`
- 33nF seems OK for the filter

```python
import matplotlib.pyplot as plt
import pandas as pd

def plotfile(filename: str) -> None:
    df = pd.read_csv(filename)
    df["ideal"] = df["current [A]"]
    ax = df.plot(x="current [A]", y="measured current [A]", style="r*-")
    df.plot(ax=ax, x="current [A]", y="ideal", style="g-")
    plt.grid()
    df.plot(ax=ax, x="current [A]", y="voltage [V]", style="b*-", secondary_y=True)
```

```python
plotfile("500mV.csv")
```

```python
plotfile("440mV.csv")
```
